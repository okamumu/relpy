```python
import numpy as np
import scipy.sparse as sp
import relpy as rp
import matplotlib.pyplot as plt
```

```python
with open('model.py', 'w') as f:
    script = rp.parse('./ex.sharpe')
    f.write(script)
```

```python
import model
t = rp.Parameter('t')
Non_Computers = model.ftree_Non_Computers(_time=t, mIRS=model.mIRS, mPRS=model.mPRS, mSA=model.mSA)
CRASH = model.ftree_CRASH(_time=t, mCS=model.mCS, Non_Computers=Non_Computers)
```

```python
model.env
```

```python
x = np.linspace(0, 10, 100)
y = [CRASH.teval(t, u, model.env) for u in x]
plt.plot(x, y)
```

```python
x = np.linspace(0, 10, 100)
y = [CRASH.tderiv(t, u, model.env, model.mIRS) for u in x]
plt.plot(x, y)
```

```python
x = np.linspace(0, 10, 100)
y = [CRASH.tderiv2(t, u, model.env, model.mIRS, model.mCS) for u in x]
plt.plot(x, y)
```
